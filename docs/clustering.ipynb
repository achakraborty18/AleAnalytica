```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
```

```python
df = pd.read_excel('../data/kantar6k_independent.xlsx')
df.head(5)
```

```python
df.fillna(df.median(), inplace=True)
df.isna().sum().sum()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
linked = linkage(scaled_data, method='ward')  # You can try different linkage methods
```

```python
# Plot the dendrogram to determine the number of clusters
plt.figure(figsize=(12, 6))
dendrogram(linked, p=8, truncate_mode='lastp', orientation='top', distance_sort='descending')
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
fig = ff.create_dendrogram(df)
fig.update_layout(width=800, height=500)
fig.show()
```
